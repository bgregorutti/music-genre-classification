```python
import librosa, numpy as np, sys, os, pandas as pd
from librosa.display import waveplot,specshow
import IPython.display
from IPython.display import Audio

import matplotlib.pyplot as plt
plt.register_cmap(name='magma', cmap=plt.get_cmap('Spectral'))
plt.set_cmap(plt.get_cmap('magma'))
```

```python
def set_figure(figsize=(12,8)):
    fig = plt.figure(figsize=figsize)
    fig.patch.set_facecolor('xkcd:white')
    return fig
```

### Load data

```python
file_1 = 'data/genres_wav/blues_00000.wav'
file_2 = 'data/genres_wav/metal_00000.wav'

y_1, sr_1 = librosa.load(file_1)
m, s = divmod(len(y_1)/sr_1, 60)
print('Done. len(y_1) = {}. Signal rate = {} Hz. Time: {}.{}s'.format(len(y_1),sr_1,m,s))

y_2, sr_2 = librosa.load(file_2)
m, s = divmod(len(y_2)/sr_2, 60)
print('Done. len(y_2) = {}. Signal rate = {} Hz. Time: {}.{}s'.format(len(y_2),sr_2,m,s))
```

```python
S_1 = librosa.stft(y_1)
log_S_1 = librosa.amplitude_to_db(S_1, ref=np.max)
print(log_S_1.shape)

S_2 = librosa.stft(y_2)
log_S_2 = librosa.amplitude_to_db(S_2, ref=np.max)
print(log_S_2.shape)
```

```python
set_figure((16, 4))
specshow(log_S_1, y_axis='log', x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Blues')
plt.tight_layout()

set_figure((16, 4))
specshow(log_S_2, y_axis='log', x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Metal')
plt.tight_layout()
```
